```python
import pandas as pd
```

```python
print("May take a few seconds!")

csv_path = "../MTA_Subway_Hourly_Ridership_Jan_Through_May_2025.csv"

df = pd.read_csv(csv_path, low_memory=False)
```

```python
# identify unique station/lat/long entries
unique_station_complex_latlong = df[['station_complex', 'station_complex_id', 'latitude', 'longitude']].drop_duplicates().reset_index(drop= True)

# number of lat/long entries per station
latlongs_per_complex = unique_station_complex_latlong.groupby(['station_complex', 'station_complex_id']).size().reset_index()

# culprits
more_than_one_latlong = latlongs_per_complex[latlongs_per_complex[0] > 1].sort_values(by = 0, ascending = False).rename(columns={0: 'num_unique_lat_long_pairs'}).reset_index(drop=True)

more_than_one_latlong
```
